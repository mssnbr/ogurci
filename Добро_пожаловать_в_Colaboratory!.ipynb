```
class Monster:
    def __init__(self, damage, name, special_attack, HP, MP):
        self.damage = damage
        self.name = name
        self.special_attack = special_attack
        self.HP = HP
        self.MP = MP

    def attack(self, player):
        print(f'{self.name} атакует. Урон: {self.damage}')
        player.HP -= self.damage

    def use_special_attack(self, player):
        if self.MP >= 50:
            print(f'{self.name} использует специальную атаку. Урон: {self.special_attack}')
            player.HP -= self.special_attack
            self.MP -= 50
        else:
            print(f'{self.name} не хватает MP для специальной атаки!')


class Player:
    def __init__(self, damage, name, special_attack, HP, weapon):
        self.damage = damage
        self.name = name
        self.special_attack = special_attack
        self.HP = HP
        self.weapon = weapon
        self.MP = 150

    def attack(self, monster):
        total_damage = self.damage + self.weapon.damage
        print(f'{self.name} атакует. Урон: {total_damage}')
        return total_damage

    def use_special_attack(self, monster):
        if self.MP >= 50:
            print(f'{self.name} использует специальную атаку. Урон: {self.special_attack}')
            monster.HP -= self.special_attack
            self.MP -= 50
            return self.special_attack
        else:
            print(f'{self.name} не хватает MP для специальной атаки!')
            return 0

    def treatment(self):
        if self.MP >= 10:
            print(f'{self.name} лечится. Восстановливает 4 HP')
            self.HP += 4
            self.MP -= 10
        else:
            print(f'{self.name} не хватает MP для лечения!')


class Weapon:
    def __init__(self, damage, name):
        self.damage = damage
        self.name = name



Ogr = Monster(3, 'Огр', 5, 30, 100)
Skelet = Monster(2, 'Скелет', 5, 15, 100)


Name_player = input('Введите имя персонажа: ')
print('Классы:')
print('Воин - 1')
print('Лучник - 2')
class_a = input('Введите номер класса:')
print('Оружие: ')
print('Меч - 1')
print('Лук - 2')
a = input('Оружие: ')
if class_a == 1:
  weapon_1 = Weapon(3, 'Меч')
  weapon_2 = Weapon(1, 'Лук')
elif class_a == 2:
  weapon_1 = Weapon(1, 'Меч')
  weapon_2 = Weapon(3, 'Лук')
if a == '1':
    weapon = weapon_1
    player_1 = Player(1, Name_player, 6, 20, weapon)
elif a == '2':
    weapon = weapon_2
    player_1 = Player(1, Name_player, 6, 20, weapon)

print('Урон игрока(без оружия):', player_1.damage)
print('Имя игрока:', player_1.name)
print('Специальная атака:', player_1.special_attack)
print('HP игрока:', player_1.HP)
if class_a == 1:
  print('Класс игрока: Воин')
elif class_a == 2:
  print('Класс игрока: Лучник')


print("Бой со Огром!")
while Ogr.HP > 0 and player_1.HP > 0:
    print('Удар - 1')
    print('Лечение - 2')
    print('Специальная атака - 3')
    print('Мана игрока:',player_1.MP)
    fight = int(input('Выберите (номер) удар, лечение или супер удар: '))
    if fight == 1:
        damage = player_1.attack(Ogr)
        Ogr.HP -= damage
    elif fight == 2:
        player_1.treatment()
        print(f'HP игрока после лечения: {player_1.HP}')
    elif fight == 3:
        player_1.use_special_attack(Ogr)
    print('-------------------------')

    if Ogr.HP > 0 and Ogr.HP > 5:
        print(f'HP Огра: {Ogr.HP}')
        Ogr.attack(player_1)
        print(f'HP игрока после атаки Огра: {player_1.HP}')
    elif Ogr.HP > 0 and Ogr.HP < 5:
        print(f'HP Огра: {Ogr.HP}')
        Ogr.use_special_attack(player_1)
        print(f'HP игрока после атаки Огра: {player_1.HP}')

    print('-------------------------')

if Ogr.HP <= 0:
    print(f'{Ogr.name} повержен!')
elif player_1.HP <= 0:
    print(f'{player_1.name} пал в бою!')


if a == '1':
    weapon = weapon_1
    player_1 = Player(1, Name_player, 6, 20, weapon)
elif a == '2':
    weapon = weapon_2
    player_1 = Player(1, Name_player, 6, 20, weapon)

print("Бой со Скелетом!")
while Skelet.HP > 0 and player_1.HP > 0:
    print('Удар - 1')
    print('Лечение - 2')
    print('Специальная атака - 3')
    print('Мана игрока:',player_1.MP)
    fight = int(input('Выберите (номер) удар, лечение или супер удар: '))
    if fight == 1:
        damage = player_1.attack(Skelet)
        Skelet.HP -= damage
    elif fight == 2:
        player_1.treatment()
        print(f'HP игрока после лечения: {player_1.HP}')
    elif fight == 3:
        player_1.use_special_attack(Skelet)
    print('-------------------------')

    if Skelet.HP > 0 and Skelet.HP > 5:
        print(f'HP Скелета: {Skelet.HP}')
        Skelet.attack(player_1)
        print(f'HP игрока после атаки Скелета: {player_1.HP}')
    elif Skelet.HP > 0 and Skelet.HP < 5:
        print(f'HP Скелета: {Skelet.HP}')
        Skelet.use_special_attack(player_1)
        print(f'HP игрока после атаки Скелета: {player_1.HP}')
    print('-------------------------')

if Skelet.HP <= 0:
    print(f'{Skelet.name} повержен!')
elif player_1.HP <= 0:
    print(f'{player_1.name} пал в бою!')
```
